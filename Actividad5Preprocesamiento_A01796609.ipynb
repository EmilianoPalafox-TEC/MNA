5. ¿Tendrán los valores faltantes de `Weight` relación con alguna otra variable? Analízalo considerando las siguientes columnas:
*   Numéricas: `Age, Number of siblings`
*   Categóricas: `Gender, Education`

Para ello deberás dividir cada columna en dos grupos: con/sin faltantes en `Weight` y utilizar métodos gráficos y la prueba t para ver si hay diferencias significativas. NO olvides incluir las conclusiones del análisis con los resultados obtenidos.

**NOTA**. Puedes reutilizar las funciones `Diagnose_MV_Numerical` y `Diagnose_MV_Categorical` definidas en Hands-On-Prep_Missing_Values.zip. Revisa nuevamente cómo utilizarlas.

```python
from scipy.stats import ttest_ind
def Diagnose_MV_Numerical(df,str_att_name,BM_MV):
    MV_labels = {True:'With Missing Values',False:'Without Missing Values'}

    labels=[]
    box_sr = pd.Series('',index = BM_MV.unique())
    for poss in BM_MV.unique():
        BM = BM_MV == poss
        box_sr[poss] = df[BM][str_att_name].dropna()
        labels.append(MV_labels[poss])

    plt.boxplot(box_sr,vert=False)
    plt.yticks([1,2],labels)
    plt.xlabel(str_att_name)
    plt.show()

    plt.figure(figsize=(10,4))

    att_range = (df[str_att_name].min(),df[str_att_name].max())

    for i,poss in enumerate(BM_MV.unique()):
        plt.subplot(1,2,i+1)
        BM = BM_MV == poss
        df[BM][str_att_name].hist()
        plt.xlim = att_range
        plt.xlabel(str_att_name)
        plt.title(MV_labels[poss])

    plt.show()

    group_1_data = df[BM_MV][str_att_name].dropna()
    group_2_data = df[~BM_MV][str_att_name].dropna()

    p_value = ttest_ind(group_1_data,group_2_data).pvalue

    print('p-value of t-test: {}'.format(p_value))
```

```python
numerical_attributes = ['Age', 'Number of siblings']

BM_MV = response_df.Weight.isna()

for att in numerical_attributes:
    print('Diagnosis Analysis of Missing Values for {}:'.format(att))
    Diagnose_MV_Numerical(response_df,att,BM_MV)
    print('- - - - - - - - - - - - divider - - - - - - - - - - - ')
```

La conclusiones que podemos dar son las siguientes:


*   Boxplots: Mirando los diagramas de caja, podemos ver que el valor del `Weight` no cambia significativamente entre las dos poblaciones, sin los Missing Values las bigotes del Boxplot se amplían, así como se logran capturar algunos outliers, sin embargo, la media y la caja sigue siendo similares, lo que demuestra que ni el cambio de *Age* ni el de *Number of Siblings*  podría haber causado o influenciado la aparición de valores faltantes en Weight.
*   Histograma: El histograma de *Age* presenta una distribución similar en ambos casos, resuena un poco el histograma de *Number of Siblings*, pues la versión del historgrama con Valores Faltantes si muestra una distribución algo diferente.
* P-Value: El p-value fue de:
  1.   *Age*: p-value of t-test: 0.7258243979822129
  2.   *Number of Sibligns*: p-value of t-test: 0.8255092107556896

Ambos p-values son mayores a 0.05, por lo que no tenemos suficiente evidencia para rechazar la hipótesis nula (H₀).
Esto significa que no detectamos diferencias significativas entre los dos grupos y por lo tanto, los valores faltantes en Weight no están relacionados con las variables *Age* y *Number of Siblings*.





```python
from scipy.stats import chi2_contingency
def Diagnose_MV_Categorical(df,str_att_name,BM_MV):
    MV_labels = {True:'With Missing Values',False:'Without Missing Values'}

    plt.figure(figsize=(10,4))
    for i,poss in enumerate(BM_MV.unique()):
        plt.subplot(1,2,i+1)
        BM = BM_MV == poss
        df[BM][str_att_name].value_counts().plot.bar()
        plt.title(MV_labels[poss])
    plt.show()

    contigency_table = pd.crosstab(BM_MV,df[str_att_name])
    p_value = chi2_contingency(contigency_table)[1]

    print('p-value of Chi_squared test: {}'.format(p_value))
```

```python
categorical_attributes = ['Gender', 'Education']

BM_MV = response_df.Weight.isna()
for att in categorical_attributes:
    print('Diagnosis Analysis of Missing Values for {}:'.format(att))
    Diagnose_MV_Categorical(response_df,att,BM_MV)
    print('- - - - - - - - - - - - divider - - - - - - - - - - - ')
```

La conclusiones que podemos dar son las siguientes:

*   Gráfico de barras: Al observar los gráficos de barras nos damos cuenta de que los valores faltantes siguen el mismo comportamiento que el dataframe sin valores faltantes, por lo que no existe una tendencia significativa para creer que haya una razón sistemática para que los valores faltantes ocurran debido a un cambio en el valor de *Gender* o *Education*.
* P-Value: El p-value fue de:
  1.   *Gender*: p-value of t-test: 0.5472167199163878
  2.   *Education*: p-value of t-test: 0.8424822989541867

Ambos p-values son mayores a 0.05, por lo que no tenemos suficiente evidencia para rechazar la hipótesis nula (H₀).
Esto significa que no detectamos diferencias significativas entre los dos grupos y por lo tanto, los valores faltantes en Weight no están relacionados con las variables *Gender* ni *Education*.


6. Utiliza una medida de tendencia central para imputar los valores faltantes de `Weight`. ¿Cuál es recomendable y por qué?

Recordemos que los valores de sesgo y curtosis de la columna **Weight** son:
*   Sesgo: 1.2218702827632462
*   Curtosis: 3.943249236622443

Por lo tanto, al tener una distribución asimetrica, lo recomendable es usar la mediana, ya que es una medida de tendencia central más robusta frente a la asimetría y outliers.

```python
response_df['Weight'] = response_df['Weight'].fillna(value=response_df['Weight'].median())
```

7. ¿Hay valores atípicos de la variable `Weight`? Confírmalo con un gráfico de boxplot. ¿Cuántos son? Almacénalos en un dataframe independiente y elimínalos del conjunto original.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=response_df['Weight'], showmeans=True, color="#84F1EE")
plt.title('Boxplot de la variable Weight')
plt.tight_layout()
plt.show()
```

La variable Weight presenta outliers del lado derecho.

Analizaremos los outliers mediante el método IQR debido a que la distribución no es normal.

```python
Q1 = response_df['Weight'].quantile(0.25)
Q3 = response_df['Weight'].quantile(0.75)
IQR = Q3 - Q1

lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR
```

```python
outliers_df = response_df[(response_df['Weight'] < lower_limit) | (response_df['Weight'] > upper_limit)]
print(f"Número de valores atípicos en Weight: {len(outliers_df)}")
outliers_df['Weight'].head(10)
```

```python
response_df_clean = response_df[~((response_df['Weight'] < lower_limit) | (response_df['Weight'] > upper_limit))]
print(f"Número de registros después de eliminar outliers: {len(response_df_clean)}")
response_df_clean['Weight'].head(10)
```
